```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
from nltk.corpus import stopwords
import re
import string
import re
import string
from pythainlp.tokenize import word_tokenize 
```

```python
def remove_only_thai(text):
    return re.sub(r'[^\u0E00-\u0E7F\s]', '', text)

def remove_numbers(text):
    return re.sub(r'\d+', '', text)

def remove_alone_characters(text):
    tokens = word_tokenize(text,engine='longest', keep_whitespace=False)
    filtered_tokens = [token for token in tokens if len(token) > 1]
    return ' '.join(filtered_tokens)

def tokenization_thai(text):
    token = word_tokenize(text,engine='longest', keep_whitespace=False)
    return token
```

```python
def apply_all(INPUT_TEXT2):
    INPUT_TEXT2 = remove_only_thai(INPUT_TEXT2)
    INPUT_TEXT2 = remove_numbers(INPUT_TEXT2)
    INPUT_TEXT2 = remove_alone_characters(INPUT_TEXT2)
    INPUT_TEXT2 = tokenization_thai(INPUT_TEXT2)
    return INPUT_TEXT2
```

```python
if __name__ == "__main__":
    data = []
    list_ = set()
    d_1 = apply_all(d1)
    d_2 = apply_all(d2)
    d_3 = apply_all(d3)
    d_4 = apply_all(d4)
    d_5 = apply_all(d5)
    data.append(d_1)
    data.append(d_2)
    data.append(d_3)
    data.append(d_4)
    data.append(d_5)
    list_.update(d_1)
    list_.update(d_2)       
    list_.update(d_3)
    list_.update(d_4)
    list_.update(d_5)
    list_array = list(list_)
    print(list_)
    print(data)
```

```python
result = {}
def to_binary_weight(sentence, targets):
    data_ = []
    for target in targets:
        if target in sentence:
            data_.append(1)
        else:
            data_.append(0)
    return data_

binary_output = to_binary_weight(data[0], list_array)
for item in data:
    binary_output = to_binary_weight(item, list_array)
    result["D" + str(data.index(item)+1)] = binary_output

from tabulate import tabulate

header = list_array
rows = list(result.values())

print(tabulate(rows, headers=header, tablefmt="github"))

```
